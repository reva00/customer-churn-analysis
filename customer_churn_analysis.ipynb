```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
df = df.dropna()
```

```python
df["Churn"].value_counts()
```

```python
sns.countplot(x="Churn", data=df)
plt.show()
```

```python
sns.boxplot(x="Churn", y="MonthlyCharges", data=df)
plt.show()
```

```python
df.groupby("Churn")["MonthlyCharges"].mean()
```

```python
Insight:
Customers who churn tend to have higher monthly charges compared to those who stay, indicating pricing may influence customer retention.
```
